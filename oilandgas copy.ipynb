### Cleaning Data

Mengisi Field dari Producing Formation

```python
data["Field"] = data["Field"].fillna(data["Producing Formation"])
```

Mengisi County dari Location

```python
data["County"] = data["County"].fillna(
    data["Location"].str.split(",").str[0]
)
```

Mengisi Town dari Location

```python
data["Town"] = data["Town"].fillna(
    data["Location"].str.split(",").str[0]
)
```

Mengisi Producing Formation dari Field

```python
data["Producing Formation"] = data["Producing Formation"].fillna(data["Field"])
```

Mengisi Self-use Well menjadi INACTIVE

```python
data.loc[
    (data["Active Oil Wells"] == 0) &
    (data["Active Gas Wells"] == 0) &
    (data["Inactive Oil Wells"] == 0) &
    (data["Inactive Gas Wells"] == 0),
    "Self-use Well"
] = data["Self-use Well"].fillna("INACTIVE")
```

Mengisi Self-use Well menjadi NO

```python
data.loc[
    (
        (data["Active Oil Wells"] > 0) |
        (data["Inactive Oil Wells"] > 0) |
        (data["Active Gas Wells"] > 0) |
        (data["Inactive Gas Wells"] > 0) |
        (data["Injection Wells"] > 0) |
        (data["Disposal Wells"] > 0)
    )
    & (data["Self-use Well"].isna()),
    "Self-use Well"
] = "NO"
```

Menghapus Baris yang Masih Kosong pada Field dan Producing Formation

```python
data = data.dropna(subset=["Field", "Producing Formation"])
```

Menghapus Kolom Purchaser Codes

```python
data = data.drop(columns=["Purchaser Codes"])
```

Mengubah kolom Production Date Entered menjadi tipe data datetime

```python
data["Production Date Entered"] = pd.to_datetime(data["Production Date Entered"])
```

Menghapus data duplikat

```python
data = data.drop_duplicates()
```

Mengecek info data dan statistik data setelah di cleaning

```python
data.info()
```

```python
data.describe(include='all')
```

### Exploratory Data Analysis (EDA)

```python
data.describe(include='all')
```

Tren Produksi Minyak dan Gas per Tahun

```python
data["Year"] = data["Production Date Entered"].dt.year

trend_production = data.groupby("Year")[["Oil Produced, bbl", "Gas Produced, Mcf"]].sum()

trend_production
```

County dengan Produksi Tertinggi

```python
county_production = data.groupby("County")[["Oil Produced, bbl", "Gas Produced, Mcf"]].sum()

county_production = county_production.sort_values("Oil Produced, bbl", ascending=False)

county_production.head(10)
```

Operator dengan Produksi Tertinggi

```python
operator_production = data.groupby("Operator")[["Oil Produced, bbl", "Gas Produced, Mcf"]].sum()

operator_production = operator_production.sort_values("Oil Produced, bbl", ascending=False)

operator_production.head(10)
```

Hubungan Jumlah Sumur Aktif dengan Produksi (Membuat Total Sumur Aktif)

```python
data["Total Active Wells"] = data["Active Oil Wells"] + data["Active Gas Wells"]
```

Membuat Cluster Manual

```python
def cluster_well(x):
    if x == 0:
        return "Tidak Ada Sumur Aktif"
    elif x <= 5:
        return "Sumur Rendah"
    elif x <= 20:
        return "Sumur Sedang"
    else:
        return "Sumur Tinggi"

data["Well Cluster"] = data["Total Active Wells"].apply(cluster_well)
```

Melihat Produksi

```python
cluster_production = data.groupby("Well Cluster")[["Oil Produced, bbl","Gas Produced, Mcf"]].mean()

cluster_production
```

Mengambil Latitude dan Longitude dari Location

```python

data["Latitude"] = data["Location"].str.extract(r'\((.*?),')[0]
data["Longitude"] = data["Location"].str.extract(r',\s(.*?)\)')[0]

data["Latitude"] = pd.to_numeric(data["Latitude"])
data["Longitude"] = pd.to_numeric(data["Longitude"])
```

Melihat Statistik Produksi Berdasarkan Lokasi

```python
location_production = data.groupby(["Latitude","Longitude"])[["Oil Produced, bbl","Gas Produced, Mcf"]].sum()

location_production = location_production.sort_values(by="Oil Produced, bbl", ascending=False)

location_production.head()
```

**Insight**

Berdasarkan hasil Exploratory Data Analysis (EDA), terlihat bahwa produksi minyak dan gas mengalami fluktuasi selama periode analisis. Pada awal periode (sekitar tahun 1984–1993) produksi minyak berada pada kisaran ratusan ribu barel per tahun dan produksi gas mencapai puluhan juta Mcf, kemudian terjadi lonjakan produksi yang sangat tinggi pada tahun 1994 sebelum kembali menurun secara signifikan pada tahun-tahun berikutnya hingga mencapai tingkat yang sangat rendah pada awal tahun 2000-an. Dari sisi wilayah, County Cattaraugus tercatat sebagai daerah dengan produksi minyak tertinggi, sedangkan County Chautauqua memiliki produksi gas paling besar dibandingkan wilayah lainnya, menunjukkan adanya spesialisasi sumber daya di tiap wilayah. Analisis berdasarkan operator menunjukkan bahwa Pennzoil Products Co. merupakan operator dengan produksi minyak terbesar, sementara beberapa operator lain seperti Quaker State Corp. dan Berea Oil & Gas Corp. juga memiliki kontribusi signifikan terutama pada produksi gas. Hasil clustering manual berdasarkan jumlah sumur aktif menunjukkan hubungan yang jelas antara jumlah sumur dan tingkat produksi, di mana kelompok Sumur Tinggi menghasilkan rata-rata produksi minyak dan gas paling besar, sedangkan kelompok tanpa sumur aktif memiliki produksi yang sangat rendah. Selain itu, analisis koordinat geografis menunjukkan bahwa produksi energi terkonsentrasi pada beberapa titik lokasi tertentu, dengan beberapa koordinat seperti sekitar (42.088061, -78.491258) dan (42.023289, -78.628438) menjadi hotspot produksi minyak dan gas, yang mengindikasikan adanya area eksploitasi sumber daya yang lebih intensif dibandingkan wilayah lainnya.
